```python
import pandas as pd

# List of filenames to merge
files = ['Aggregate_bundes.csv', 'Aggregate_caf.csv', 'Aggregate_copa.csv', 'Aggregate_eu.csv', 'Aggregate_wc.csv', 'Aggregate_wspl.csv', 'Aggregate_wwc.csv', 'Aggregate_epl.csv', 'Aggregate_laliga2.csv', 'Aggregate_laliga.csv', 'Aggregate_ligue1.csv', 'Aggregate_mls.csv']

# Read each file and store them in a list
df_list = [pd.read_csv(file) for file in files]

# Concatenate all dataframes into one
merged_df = pd.concat(df_list, ignore_index=True)

# Save the combined data to a new CSV file
merged_df.to_csv('Merged_Aggregate.csv', index=False)

print(f"Successfully merged {len(files)} files into 'Merged_Aggregate.csv'")
```
